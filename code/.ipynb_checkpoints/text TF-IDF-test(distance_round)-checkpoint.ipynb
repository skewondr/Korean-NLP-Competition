```python
"""edit"""
import pandas as pd
import sentencepiece as spm
sp = spm.SentencePieceProcessor()

model_name = 'm_30.model'
vocab_name = 'm_30.vocab'
sp.load('./sp_models/'+ model_name)

import numpy as np
# Load
read_dictionary = np.load('../data/token/distance_round/elephant_dict.npy' ,allow_pickle='TRUE').item()

train_data = pd.read_pickle("../data/train/train_data.pkl")
test_data = pd.read_pickle("../data/test/test_data.pkl")

seed = 100
train_data = train_data.sample(n = 30000, random_state = seed) 
test_data = test_data.sample(n = 3000, random_state = seed) 
train_data.reset_index(drop=True, inplace=True)
test_data.reset_index(drop=True, inplace=True)
```

```python
"""edit"""

from collections import Counter

def count_values(dictionary):
    numbers = []
    for d in dictionary.keys():
        numbers.append(dictionary[d])
    ret = Counter(numbers)
    return ret

def connect_keys(dictionary, sp_model):
    init_dict = count_values(dictionary)
    new_keys=[]
    for d in init_dict.keys():
        new_keys.append(sp.id_to_piece(d))
    return new_keys
```

```python
"""edit"""

from tqdm import tnrange, tqdm_notebook

def sentence_to_token(data, col_name, sp_model): #sp_model : sentencepiece model
    X_pieces=[]
    for i in tnrange(len(data)):
        X_pieces.append(sp_model.encode_as_pieces(data.loc[i,col_name]))
    return X_pieces
    
def convert_token(X_pieces, dictionary, sp_model):
    X_pieces2=[]
    for i, tokens in enumerate(X_pieces):
        token_to_token = []
        for j, token in enumerate(tokens):
            if token in dictionary: #만약 토큰이 elephant dict 안에 있는 단어라면
                token_to_token.append(sp_model.id_to_piece(dictionary[token]))
            else : token_to_token.append(token)
        X_pieces2.append(token_to_token)
    return X_pieces2
  
# def elephant_embedding(data, dictionary, sp_model, text_col, label_col):
#     X_pieces = sentence_to_token(data, text_col, sp_model)
#     X_ids = token_to_id(X_pieces, dictionary, sp_model)
#     df = pd.DataFrame(X_ids).join(data[label_col])
#     return X_pieces, X_ids, df
```

```python
#text_to_matrix 기능 구현
"""edit"""

def wordDict(sentence, keys):
    wordDict = dict.fromkeys(keys,0)
    if len(wordDict) != 13190 : print('error')
    for word in sentence:
        try:
            wordDict[word]+=1 #한 문장 내에서 단어의 등장 빈도수를 계산. 
        except KeyError:
            continue
    return wordDict
     
```

```python
# def elephant_embedding(data, dictionary, sp_model, text_col, label_col):
#     X_pieces = sentence_to_token(data, text_col, sp_model)
#     X_ids = token_to_id(X_pieces, dictionary, sp_model)
#     df = pd.DataFrame(X_ids).join(data[label_col])
#     return X_pieces, X_ids, df
"""edit"""

def computeWB(pieces, keys):
    wordBoard = pd.DataFrame()
    for index in tnrange(len(pieces)):
        newDict = wordDict(pieces[index], keys)
        wordBoard = wordBoard.append(newDict, ignore_index = True)
    return wordBoard
```

```python
train_pieces = sentence_to_token(train_data, 'Text', sp)
train_pieces2 = convert_token(train_pieces, read_dictionary, sp)
test_pieces = sentence_to_token(test_data, 'Text', sp)
test_pieces2 = convert_token(test_pieces, read_dictionary, sp)
flatList = []
for elem in train_pieces2:
    flatList.extend(elem)
keys = list(set(flatList))
```

```python
wordBoard_train = computeWB(train_pieces2, keys)
wordBoard_test = computeWB(test_pieces2, keys)
wordBoard_test
```

```python
#tf = (Frequency of the word in the document) / (Total number of words in the document)
def computetfBoard(wordBoard, data):
    tfBoard = wordBoard.copy()
    for i in tnrange(tfBoard.shape[0]):
        for j in range(tfBoard.shape[1]):
            if tfBoard.iloc[i,j]>0:
                sentence = data[i]
                tfBoard.iloc[i,j] /= float(len(sentence))
    return tfBoard
```

```python
tfBoard_train = computetfBoard(wordBoard_train, train_pieces2)
tfBoard_test = computetfBoard(wordBoard_test, test_pieces2)
tfBoard_test
```

```python
#idf (Total number of (documents))/(Number of (documents) containing the word)
import math 

def computeIDF(wordBoard, keys):
    N = wordBoard.shape[0]
    idfDict = dict.fromkeys(keys,0)
    for i in tnrange(wordBoard.shape[0]):
        for j in range(wordBoard.shape[1]):
                if wordBoard.iloc[i,j]>0:
                    word = wordBoard.columns[j]
                    idfDict[word] +=1
    for word, val in idfDict.items():
        if val>0:
            idfDict[word] = math.log10(N/float(val))
    return idfDict
```

```python
idfs_train = computeIDF(wordBoard_train, keys)
idfs_test = computeIDF(wordBoard_test, keys)
idfs_test
```

```python
#tf-idf
def computeTFIDF(tfBoard, idfs):
    tfidfBoard = tfBoard.copy()
    for i in tnrange(tfidfBoard.shape[0]):
        for j in range(tfidfBoard.shape[1]):
            if tfidfBoard.iloc[i,j] == 0 : continue
            word = tfidfBoard.columns[j]
            tfidfBoard.iloc[i,j] *= idfs[word]
            if idfs[word] == 0 : continue
    return tfidfBoard
    
```

```python
tfidfs_train = computeTFIDF(tfBoard_train, idfs_train)
tfidfs_test = computeTFIDF(tfBoard_test, idfs_test)
tfidfs_test
```

```python
tfidfs_train=pd.concat([tfidfs_train, train_data['Label']], axis=1)
tfidfs_test=pd.concat([tfidfs_test, test_data['Label']], axis=1)
tfidfs_test
```

# SVM 

```python
from sklearn.svm import SVC
import matplotlib.pyplot as plt

c = [0.001, 0.01, 0.1, 10, 25, 50, 100, 1000]
#svm 학습 모델 함수 
def svm_func(k, x_train, y_train, x_val, y_val):
    train_scores=[]
    valid_scores=[]
    
    svm=SVC(gamma='auto',random_state=101)
    svm.kernel = k
    
    for i in c:
        svm.C = i
        svm.fit(x_train, y_train)
        
        train_score=svm.score(x_train, y_train)
        train_scores.append(train_score)
        
        valid_score=svm.score(x_val,y_val)
        valid_scores.append(valid_score)
        
    return train_scores, valid_scores, svm

def svm_visual(title, t_s, v_s):
    c=['0.001', '0.01', '0.1', '10', '25', '50', '100', '1000']
    plt.plot(c, t_s, color='orange', marker='o', label='Training Accuracy')
    plt.plot(c, v_s, color='pink', marker='o', label='Validation Accuracy')
    plt.xticks(c)
    plt.legend()
    plt.title(title)
    plt.xlabel('Value of C')
    plt.ylabel('Accuracy Score')
    plt.ylim(0.2, 1.3)
    plt.savefig('tfidf '+title+'.png', bbox_inches='tight')
    plt.show()
    
    
```

```python
seed=100
train_tfidf2 = tfidfs_train.copy()
df0=tfidfs_train.loc[(tfidfs_train['Label']==0)].sample(n = 1500, random_state = seed) 
train_tfidf2 = train_tfidf2.drop(df0.index)
print(len(train_tfidf2))
df1=tfidfs_train.loc[(tfidfs_train['Label']==1)].sample(n = 1500, random_state = seed) 
train_tfidf2 = train_tfidf2.drop(df1.index)
print(len(train_tfidf2))

train_tfidf2.reset_index(drop=True, inplace=True)
eval_tfidf = pd.DataFrame()
eval_tfidf = df0.append(df1)
eval_tfidf.reset_index(drop=True, inplace=True)
eval_tfidf
```

```python
def countLabel(data):
    count0 = 0
    count1 = 0
    for label in data:
        if label == 0: 
            count0+=1
        elif label == 1: 
            count1+=1
    total = len(data)
    print("총 데이터 개수:", total)
    print("class 0 개수: %d, 비율: %.2f"%(count0, count0/total))
    print("class 1 개수: %d, 비율: %.2f"%(count1, count1/total))
```

```python
countLabel(train_tfidf2['Label'])
countLabel(eval_tfidf['Label'])
countLabel(tfidfs_test['Label'])
```

```python


t_s, v_s, svm1 = svm_func('rbf', train_tfidf2.iloc[:,:-1], train_tfidf2['Label']
                          , eval_tfidf.iloc[:,:-1], eval_tfidf['Label'])
svm_visual('Kernel = rbf', t_s, v_s)
t_s2, v_s2, svm2 = svm_func("linear", train_tfidf2.iloc[:,:-1], train_tfidf2['Label']
                          , eval_tfidf.iloc[:,:-1], eval_tfidf['Label'])
svm_visual('Kernel = linear', t_s2, v_s2)
t_s3, v_s3, svm3 = svm_func("sigmoid", train_tfidf2.iloc[:,:-1], train_tfidf2['Label']
                          , eval_tfidf.iloc[:,:-1], eval_tfidf['Label'])
svm_visual('Kernel = sigmoid', t_s3, v_s3)

```

```python
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
#linear
svm2.C=10
svm2.fit(train_tfidf2.iloc[:,:-1], train_tfidf2['Label'])
y_pred1 = svm2.predict(tfidfs_test.iloc[:,:-1])
target_names=['0', '1']
print("Accuracy:", metrics.accuracy_score(tfidfs_test['Label'], y_pred1))
print(confusion_matrix(tfidfs_test['Label'], y_pred1))
print(classification_report(tfidfs_test['Label'], y_pred1, target_names=target_names))
```

```python
import seaborn as sn
data_= {'y_Actual': tfidfs_test['Label'], 'y_Predicted': y_pred1}
df_cm = pd.DataFrame(data_, columns=['y_Actual','y_Predicted'])
confusion_matrix = pd.crosstab(df_cm['y_Actual'], df_cm['y_Predicted'],
                               rownames=['Actual'], colnames=['Predicted'])

sn.heatmap(confusion_matrix, annot=True)
plt.savefig('tfidf_dr.png', box_inches='tight')
plt.show()
```
